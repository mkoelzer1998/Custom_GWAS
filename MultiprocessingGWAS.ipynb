```python

def chrom(chrom_num, vcf_pd, startSamples):
    global BP
    global CHR
    global SNP
    global Xj
    Xj_f = []
    BP_f = vcf_pd[vcf_pd[0] == chrom_num][1].to_numpy()
    CHR_f = vcf_pd[vcf_pd[0] == chrom_num][0].to_numpy()
    SNP_f = vcf_pd[vcf_pd[0] == chrom_num][2].to_numpy()
    vcf_arr = vcf_pd[vcf_pd[0]==chrom_num].to_numpy()
    for j in range(len(vcf_arr)):
        all_count = []
        for allele in vcf_arr[j][startSamples:len(vcf_arr[0])]:
            all_count.append(int(allele[0] != '0') + int(allele[2] != '0'))
        Xj_f.append(all_count)
        
    BP[chrom_num-1] = BP_f
    CHR[chrom_num-1] = CHR_f
    SNP[chrom_num-1] = SNP_f
    Xj[chrom_num-1] = Xj_f
```
